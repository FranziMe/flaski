```python
from matplotlib_venn import venn2, venn3, venn2_circles, venn3_circles
import matplotlib.pylab as plt
```

```python
A=["A","B","C","D","E","F"]
B=["C","D","E","F","G","H","I","J","K"]
C=["E","F","G","H","I","J","K","L","M","N"]

fig, axes = plt.subplots(1, 1)
venn_=venn2([set(A),set(B)],("label A","label B"),set_colors=('r', 'b'),alpha=0.4,normalize_to=1.0, ax=axes)

axes
```

```python
venn_=venn2([set(A),set(B)],("label A","label B"),set_colors=('r', 'b'),alpha=0.4,normalize_to=1.0)
venn=venn2_circles([set(A),set(B)],alpha=0.2, color='black', linestyle='solid', linewidth=1.0)
venn[0].set_lw(5.0)
venn[1].set_lw(8.0)
venn[1].set_color("red")
```

```python
venn_=venn3([set(A),set(B),set(C) ],("label A","label B","label C"),set_colors=('r', 'b',"y"),alpha=0.4,normalize_to=1.0)
venn=venn3_circles([set(A),set(B),set(C)],alpha=0.2, color='black', linestyle='solid', linewidth=1.0)
venn[0].set_lw(5.0)
venn[1].set_lw(8.0)
venn[2].set_lw(10.0)

venn[0].set_linestyle("dashed")
venn[0].set_alpha(1)

venn[1].set_color("red")
```
