```python
import pandas as pd
```

```python
david_df=pd.read_excel(david_file)
ge_df=pd.read_excel(gene_expression_file)
```

```python
david_df=david_df.drop(["genes name","log2fc"],axis=1)
david_df.head()
```

```python
gedic=ge_df[["ensembl_gene_id","log2FoldChange"]]
gedic.loc[:,"ensembl_gene_id"]=gedic.loc[:,"ensembl_gene_id"].apply(lambda x: x.upper() )
gedic.index=ge_df["ensembl_gene_id"].tolist()
gedic=gedic.to_dict()["log2FoldChange"]
```

```python
david_df=david_df[0:20]
david_df["-log10(p value)"]=david_df["ease"].apply(lambda x: np.log10(x)*-1)
david_df.head()
```

```python
plotdf=pd.DataFrame()
for term in david_df["termName"].tolist():
    tmp=david_df[david_df["termName"]==term]
    pvalue=tmp.iloc[0,tmp.columns.tolist().index("ease")]
    log10p=tmp.iloc[0,tmp.columns.tolist().index("-log10(p value)")]
    genes=tmp.iloc[0,tmp.columns.tolist().index("geneIds")].split(", ")
    tmp=pd.DataFrame({"term":term,"genes":genes})
    tmp["expression"]=tmp["genes"].apply(lambda x: gedic.get(x.upper()) )
    tmp=tmp.sort_values(by=["expression"], ascending=True)
    tmp.reset_index(inplace=True, drop=True)
    frac=log10p/len(genes)
    tmp["-log10(p value)"]=frac
    if pvalue < 0.001:
        pvalue="{:.3e}".format(pvalue)
    tmp["term p value"]=pvalue
    #print(log10p,len(genes),frac,tmp["-log10(p value)"].tolist()[-1])
    plotdf=pd.concat([plotdf,tmp])
plotdf.reset_index(inplace=True, drop=True)
```

```python
expression=plotdf["expression"].tolist()
low=np.percentile(expression,5)
high=np.percentile(expression, 100-5)
#plotdf.loc[plotdf["expression"]<low,"expression"]=low
#plotdf.loc[plotdf["expression"]>high,"expression"]=high
```

```python
fig = px.bar(plotdf, y="term", x="-log10(p value)", color="expression", orientation="h",
             color_continuous_scale='Bluered', \
             hover_name="genes", hover_data=["expression", "term p value"],\
            title="Cell plot",\
            range_color=[low,high],\
            width=None, height=None)

fig.update_layout({"yaxis":{"tickfont":{"size":10}}, "xaxis":{"tickfont":{"size":10}} })

fig.update_layout(title={"font":{"size":20}})

fig.update_layout(coloraxis_colorbar=dict(
    title={'text': 'Color bar', "font":{"size": 10 }},
    thicknessmode="pixels",
    ticks="outside",
    tickwidth=2,
    tickfont={"size":10},
    ticklen=5
))


# fig.update_layout(yaxis={'domain': [0, 1-float(pa["col_dendogram_ratio"]) ],
#                                     'mirror': False,
#                                     'showgrid': False,
#                                     'showline': False,
#                                     'zeroline': False,
#                                     'showticklabels': pa_["xticklabels"],
#                                     "tickfont":{"size":float(pa["xaxis_font_size"])} ,
#                                     'ticks': "",\
#                                     'tickvals':heatmap[0]['y'],\
#                                     'ticktext':dendro_leaves_x_labels})
#             colorbar={"title":{"text":"colorbar" ,"font":{"size": 3 }},
#                       "tickfont":{"size":3}}

## add gene names to over data

# fig.update_layout(hovertemplate =
#                 '<b>Term</b>: %{x}<br>'+
#                 '<br><b>Term p value</b>: %{term p value}<br>'+
#                 '<br><b>Gene</b>: %{genes}<br>'+
#                 '<br><i>Expression</i>: $%{color:.2f}<br>')
# title=None, template=None, width=None, height=None, range_color=None
# labels: dict with str keys and str values (default `{}`)
#         By default, column names are used in the figure for axis titles, legend
#         entries and hovers. This parameter allows this to be overridden. The
#         keys of this dict should correspond to column names, and the values
#         should correspond to the desired label to be displayed.
#  range_color: list of two numbers
#         If provided, overrides auto-scaling on the continuous color scale.
fig["data"][0]['hovertemplate']='<b>%{hovertext}</b><br>expression=%{marker.color}<br><br>term: %{y}<br>term p value=%{customdata[1]}'
fig.update_layout(template='plotly_white')
fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_yaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
fig.update_xaxes(ticks="outside", tickwidth=2, tickcolor='black', ticklen=5)
fig.show()
```
